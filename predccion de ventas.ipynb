```python
from statistics import mode

from matplotlib.pyplot import xlabel


def sales_predictions(store_ID,sales_df, periods):
    sales_df=sales_df[sales_df['Store']==store_ID]
    sales_df=sales_df[['Date','Sales']].rename(columns={'Date':'ds','Sales':'y'})
    sales_df=sales_df.sort_values('ds')

    model= Prophet()
    model.fit(sales_df)
    future = model.make_future_dataframe(periods=periods)
    forestcast=model.predict(future)
    figure = model.plot(forestcast, xlabel="fecha", ylabel="Ventas")
    figure2=model.plot_components(forestcast)
```

predicciones

```python
def sales_predictions_vacations(store_ID,sales_df,holidays,periods):
    sales_df=sales_df[sales_df['Store']==store_ID]
    sales_df=sales_df[['Date','Sales']].rename(columns={'Date':'ds','Sales':'y'})
    sales_df=sales_df.sort_values('ds')

    model= Prophet(holidays=holidays)
    model.fit(sales_df)
    future = model.make_future_dataframe(periods=periods)
    forestcast=model.predict(future)
    figure = model.plot(forestcast, xlabel="fecha", ylabel="Ventas")
    figure2=model.plot_components(forestcast)
```

```python
sales_predictions_vacations(6,sales_train_all_df,school_satate_holidays,90)
```
